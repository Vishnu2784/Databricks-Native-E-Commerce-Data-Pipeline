```python
import json
import random
import time
from datetime import datetime, timedelta

def generate_Streame_api(total_records=1000000, batch_size=1000):
    actions = ["view_item", "add_to_cart", "purchase", "remove_from_cart"]
    device_types = ["mobile", "desktop", "tablet"]
    for offset in range(0, total_records, batch_size):
        batch=[]
        for _ in range(batch_size):
            record = {
                "event_id": f"EVT-{random.randint(100000, 999999)}",
                "event_timestamp": (datetime.utcnow() - timedelta(minutes=random.randint(0, 10000))).isoformat(),
                "event_details": {"action": random.choice(actions), "device": random.choice(device_types)},
                "user_data": {"user_id": random.randint(1000, 5000)},
                "transaction": {"item_id": f"ITEM_{random.randint(1, 500)}", "price": round(random.uniform(10.0, 500.0), 2)}
            }
            batch.append(record)
        yield batch
        
dbfs_path = "/Volumes/project_streame/bronze_layer/batches/"

batch_count = 0
for batch in generate_Streame_api():
    batch_count += 1
    file_name = f"{dbfs_path}batch_{batch_count}_{int(time.time())}.json"
    with open(file_name, 'w') as f:
        for record in batch:
            f.write(json.dumps(record) + '\n')
            
print(f"Success! 5 JSON batch files generated in {dbfs_path}")
    
```
